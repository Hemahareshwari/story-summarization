Importing the libraries

```
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

Step by step for text summarization
1. Removing unnecessary words
2. Tokenizing the sentence 
3. Word frequency table
4. Clustering
5. Summarization

```
stopwords= list(STOP_WORDS)
nlp=spacy.load("en_core_web_sm")
```

```
doc=nlp(text)
```

```
### Tokenizing the text 
tokens=[token.text for token in doc]
tokens
```

```
punctuation=punctuation + '\n'
punctuation
```

```
word_frequecies={}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_frequecies.keys():
        word_frequecies[word.text]=1
      else:
        word_frequecies[word.text]+=1
```

```
print(word_frequecies)
```

```
max_frequency= max(word_frequecies.values())
max_frequency
```

```
for word in word_frequecies.keys():
  word_frequecies[word]=word_frequecies[word]/max_frequency
print(word_frequecies)
```

```
##sentence tocknization
sentence_tokens=[sent for sent in doc.sents]
print(sentence_tokens)
```

```
sentence_scores={}
for sent in sentence_tokens:
  for word in sent:
    if word.text.lower() in word_frequecies.keys():
      if sent not in sentence_scores.keys():
        sentence_scores[sent]=word_frequecies[word.text.lower()]
      else:
        sentence_scores[sent]+=word_frequecies[word.text.lower()]
        
      
```

```
print(sentence_scores)
```

```
from heapq import nlargest
```

```
select_length= int(len(sentence_tokens)*0.3)
select_length
```

```
### Storing sentences into our summary. 
summary=nlargest(select_length,sentence_scores,key=sentence_scores.get)
summary
```

```
final_summary=[word.text for word in summary]
summary=" ".join(final_summary)
print(summary)
```
